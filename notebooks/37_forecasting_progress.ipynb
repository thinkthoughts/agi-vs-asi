# 37 — Forecasting Progress

This notebook translates the repo's growth pathways and constraints into forecast-relevant indicators.

Notebook 31 mapped constraints and bottlenecks.

Notebook 37 asks:

> What indicators are worth watching?

The goal is not to predict a date for ASI.

The goal is to identify observable signals that should update expectations.

```text
growth pathways

      ↓

```

This notebook is designed to run from a fresh Google Colab runtime.

## 1. Clone repo, enter repo root, and install dependencies

````python
from pathlib import Path
import os
import subprocess
import sys

REPO = "agi-vs-asi"
GITHUB_USER = "thinkthoughts"
BRANCH = "main"

CONTENT = Path("/content")
ROOT = CONTENT / REPO

if not ROOT.exists():
    url = f"https://github.com/{GITHUB_USER}/{REPO}.git"
    print(f"Cloning {url}")
    subprocess.run(
        ["git", "clone", "--branch", BRANCH, url, str(ROOT)],
        check=True,
    )
else:
    print(f"Repo already exists: {ROOT}")

os.chdir(ROOT)

subprocess.run(
    [sys.executable, "-m", "pip", "install", "-q", "pyyaml", "pandas"],
    check=True,
)

print("Working directory:", Path.cwd())
````

## 2. Load `paper.yaml` and `statement.yaml`

````python
import yaml

PAPER_YAML = Path("paper.yaml")
STATEMENT_YAML = Path("statement.yaml")

if not PAPER_YAML.exists():
    raise FileNotFoundError(
        "paper.yaml not found in the repo root. "
        "Confirm that the repository has paper.yaml at /content/agi-vs-asi/paper.yaml."
    )

if not STATEMENT_YAML.exists():
    raise FileNotFoundError(
        "statement.yaml not found in the repo root. "
        "Confirm that the repository has statement.yaml at /content/agi-vs-asi/statement.yaml."
    )

with open(PAPER_YAML, "r", encoding="utf-8") as f:
    paper = yaml.safe_load(f)

with open(STATEMENT_YAML, "r", encoding="utf-8") as f:
    statement = yaml.safe_load(f)

print(paper.get("title", "No title found"))
print(paper.get("paper", "arXiv:2606.12683"))
print()
print(statement.get("equation", "capability = growth - friction"))
````

## 6. Signal vs noise

Not every impressive observation is a strong forecast signal.

The strongest signals should be:

- measured,
- replicated,
- capability-relevant,
- hard to fake,
- connected to real tasks or costs.

````python
signal_noise = pd.DataFrame(
    [
        [
            "Bigger model",
            "Weak to medium",
            "Size alone does not guarantee capability or efficiency.",
        ],
        [
            "Better benchmark score",
            "Medium",
            "Useful only if benchmark is robust and not overfit.",
        ],
        [
            "Lower cost per capability",
            "Strong",
            "Shows efficiency, deployment feasibility, or algorithmic progress.",
        ],
        [
            "Real-world deployment success",
            "Strong",
            "Shows capability outside curated benchmarks.",
        ],
        [
            "Independent replication",
            "Strong",
            "Reduces the chance that progress is artifact or hype.",
        ],
        [
            "AI-assisted discovery adopted in production",
            "Strong",
            "Evidence that feedback lifts are improving capability-building.",
        ],
        [
            "More agents in a workflow",
            "Weak to medium",
            "Agent count matters only if output improves after coordination cost.",
        ],
        [
            "Energy or hardware bottleneck reports",
            "Medium to strong",
            "Can indicate constraints becoming dominant.",
        ],
    ],
    columns=["Observation", "Forecast signal strength", "Reason"],
)

signal_noise
````

## 12. Write report

````python
reports = Path("reports")
reports.mkdir(exist_ok=True)

out = reports / "37_forecasting_progress.md"

lines = [
    "# 37 — Forecasting Progress",
    "",
    "Forecasting AGI-to-ASI progress should track indicators rather than predict one fixed date.",
    "",
    "## Forecast-relevant indicators",
    "",
]

for _, row in indicators.iterrows():
    lines.append(
        f"- **{row['Area']}:** {row['Indicator']} — {row['Why it matters']} "
        f"(difficulty: {row['Measurement difficulty']})."
    )

lines.extend([
    "",
    "## Scenarios",
    "",
])

for _, row in scenarios.iterrows():
    lines.append(f"- **{row['Scenario']}:** {row['Driver']} — {row['Interpretation']}")

lines.extend([
    "",
    "## Signal vs noise",
    "",
])

for _, row in signal_noise.iterrows():
    lines.append(
        f"- **{row['Observation']}:** {row['Forecast signal strength']} signal — {row['Reason']}"
    )

lines.extend([
    "",
    "## Dashboard candidates",
    "",
])

for _, row in dashboard_candidates.iterrows():
    lines.append(f"- **{row['Dashboard area']}:** {row['Candidate signals']}")

lines.extend([
    "",
    "## Development questions",
    "",
])

for i, question in enumerate(questions, start=1):
    lines.append(f"{i}. {question}")

lines.extend([
    "",
    "## Working statement",
    "",
    "Forecasting progress means tracking which growth pathways are accelerating and which constraints are becoming dominant.",
])

out.write_text("\n".join(lines), encoding="utf-8")

print("Wrote", out)
````
