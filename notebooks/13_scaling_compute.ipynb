# 13 — Scaling Compute

This notebook connects the AGI → ASI transition to the scaling pathway discussed in arXiv:2606.12683.

Notebook 07 identified digital advantages:

- speed,
- memory,
- replication,
- sharing,
- parallelism.

Notebook 13 asks:

> What happens when those advantages are supplied with more effective compute?

The focus is not prediction certainty. The focus is constraint mapping:

```text
effective compute growth
        ↓
capability growth
        ↓
frictions and bottlenecks
```

This notebook is designed to run from a fresh Google Colab runtime.

## 1. Clone repo, enter repo root, and install dependencies

````python
from pathlib import Path
import os
import subprocess
import sys

REPO = "agi-vs-asi"
GITHUB_USER = "thinkthoughts"
BRANCH = "main"

CONTENT = Path("/content")
ROOT = CONTENT / REPO

if not ROOT.exists():
    url = f"https://github.com/{GITHUB_USER}/{REPO}.git"
    print(f"Cloning {url}")
    subprocess.run(
        ["git", "clone", "--branch", BRANCH, url, str(ROOT)],
        check=True,
    )
else:
    print(f"Repo already exists: {ROOT}")

os.chdir(ROOT)

subprocess.run(
    [sys.executable, "-m", "pip", "install", "-q", "pyyaml", "pandas"],
    check=True,
)

print("Working directory:", Path.cwd())
````

## 2. Load `paper.yaml`

````python
import yaml

PAPER_YAML = Path("paper.yaml")

if not PAPER_YAML.exists():
    raise FileNotFoundError(
        "paper.yaml not found in the repo root. "
        "Confirm that the repository has paper.yaml at /content/agi-vs-asi/paper.yaml."
    )

with open(PAPER_YAML, "r", encoding="utf-8") as f:
    data = yaml.safe_load(f)

print(data.get("title", "No title found"))
print(data.get("paper", "arXiv:2606.12683"))
````

## 3. Core question

The scaling pathway asks whether increases in effective compute can continue to produce meaningful capability growth.

The paper frames effective compute as a combination of several forces:

```text
hardware improvement
× investment growth
× algorithmic efficiency
≈ effective compute growth
```

The important point is that scaling is not just “bigger models.”

It includes training compute, inference compute, data, hardware, energy, and algorithmic efficiency.

````python
core_question = (
    "Can scaling compute, models, data, and algorithmic efficiency sustain "
    "capability growth from AGI toward ASI?"
)

print(core_question)
````

## 4. Scaling components

This table defines what “scaling” means in this repo.

````python
import pandas as pd

scaling_components = pd.DataFrame(
    [
        [
            "Hardware improvement",
            "More computation per dollar or per watt",
            "Expands feasible training and inference budgets",
        ],
        [
            "Compute investment",
            "More chips, clusters, datacenters, and energy infrastructure",
            "Expands total available compute stock",
        ],
        [
            "Algorithmic efficiency",
            "Less compute needed to reach a capability level",
            "Acts like extra hardware without physically adding hardware",
        ],
        [
            "Training scale",
            "Larger or longer training runs",
            "Can improve model capability if returns continue",
        ],
        [
            "Inference scale",
            "More test-time search, reasoning, sampling, and agent operation",
            "Can improve performance after training",
        ],
        [
            "Data scale and quality",
            "More or better training and interaction data",
            "Can support generalization if quality remains high",
        ],
    ],
    columns=["Component", "Meaning", "Why it matters"],
)

scaling_components
````

## 5. Effective compute frame

A simple repo-level abstraction:

```text
effective compute = hardware × investment × algorithmic efficiency
```

This is not meant as a precise fitted model.

It is a bookkeeping frame for separating sources of growth.

````python
effective_compute_factors = pd.DataFrame(
    [
        ["Hardware", "physical compute per dollar / watt", "chips, memory, networking, fabrication"],
        ["Investment", "total deployed compute", "datacenters, capital, energy procurement"],
        ["Algorithms", "capability per unit compute", "architecture, optimization, data efficiency"],
        ["Inference", "test-time compute per task", "search, planning, tool use, agent loops"],
        ["Data", "usable learning signal", "human data, synthetic data, environment interaction"],
    ],
    columns=["Factor", "Scaling role", "Examples"],
)

effective_compute_factors
````

## 6. Scaling growth vs scaling friction

Every growth mechanism has a corresponding friction.

This is the core Climate Reality reading of the paper:

```text
capability = growth - friction
```

````python
scaling_friction = pd.DataFrame(
    [
        [
            "Hardware",
            "More raw compute",
            "chip supply, memory bandwidth, interconnect, cost, physical fabrication limits",
        ],
        [
            "Investment",
            "Larger clusters and more deployed compute",
            "capital cost, energy, permitting, grid constraints",
        ],
        [
            "Algorithms",
            "More capability per FLOP",
            "diminishing returns, benchmark overfitting, research bottlenecks",
        ],
        [
            "Training",
            "larger model or longer training",
            "data quality, instability, cost, evaluation gaps",
        ],
        [
            "Inference",
            "more thinking, search, sampling, agent steps",
            "latency, cost, reliability, compounding errors",
        ],
        [
            "Data",
            "more learning signal",
            "data exhaustion, synthetic degeneration, distribution shift",
        ],
    ],
    columns=["Scaling dimension", "Growth effect", "Friction"],
)

scaling_friction
````

## 7. Scaling and digital advantages

Notebook 07 introduced digital advantages.

Scaling compute makes those advantages operational.

````python
advantage_scaling = pd.DataFrame(
    [
        [
            "Speed",
            "More inference and reasoning steps per unit time",
            "latency, energy, serial bottlenecks",
        ],
        [
            "Memory",
            "Larger context, retrieval, and persistent state",
            "retrieval quality, coherence, stale information",
        ],
        [
            "Replication",
            "More instances can run in parallel",
            "coordination, security, duplicated effort",
        ],
        [
            "Sharing",
            "Training traces and generated data can be reused",
            "evaluation quality, privacy, alignment, contamination",
        ],
        [
            "Parallelism",
            "Agent collectives can operate at larger scale",
            "communication overhead, organization, incentives",
        ],
    ],
    columns=["Digital advantage", "How scaling amplifies it", "Friction"],
)

advantage_scaling
````

## 8. Scaling is not one question

Instead of asking:

```text
Is scaling enough?
```

this repo asks more specific questions:

```text
Which scaling dimension?
Which capability?
Which bottleneck?
Which time horizon?
Which feedback loop?
```

````python
scaling_questions = [
    "Does more training compute still improve broad capability?",
    "Does more inference compute improve reliability or only verbosity?",
    "Does algorithmic efficiency continue compounding like effective hardware growth?",
    "Does data quality become the limiting factor?",
    "Do energy and datacenter constraints dominate physical scaling?",
    "Can scaling many AGI-level agents produce ASI-level collective capability?",
]

for i, question in enumerate(scaling_questions, start=1):
    print(f"{i}. {question}")
````

## 10. Write report

````python
reports = Path("reports")
reports.mkdir(exist_ok=True)

out = reports / "13_scaling_compute.md"

lines = [
    "# 13 — Scaling Compute",
    "",
    "Scaling compute is one pathway from AGI toward ASI, but it is not a single variable.",
    "",
    "The repo tracks scaling through hardware, investment, algorithmic efficiency, training, inference, and data.",
    "",
    "## Scaling components",
    "",
]

for _, row in scaling_components.iterrows():
    lines.append(f"- **{row['Component']}:** {row['Meaning']} — {row['Why it matters']}")

lines.extend([
    "",
    "## Effective compute factors",
    "",
])

for _, row in effective_compute_factors.iterrows():
    lines.append(f"- **{row['Factor']}:** {row['Scaling role']} ({row['Examples']})")

lines.extend([
    "",
    "## Growth vs friction",
    "",
])

for _, row in scaling_friction.iterrows():
    lines.append(
        f"- **{row['Scaling dimension']}:** growth — {row['Growth effect']}; friction — {row['Friction']}."
    )

lines.extend([
    "",
    "## Scaling questions",
    "",
])

for i, question in enumerate(scaling_questions, start=1):
    lines.append(f"{i}. {question}")

lines.extend([
    "",
    "## Working statement",
    "",
    "Scaling is not only larger models. It is a coupled system of hardware, investment, algorithms, inference, data, and constraints.",
])

out.write_text("\n".join(lines), encoding="utf-8")

print("Wrote", out)
````
