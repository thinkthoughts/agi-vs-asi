# 01 — AGI vs ASI

This notebook clarifies the distinction between:

- AGI
- ASI
- Universal AI

and establishes the conceptual foundation for the rest of the `agi-vs-asi` repository.

It is designed to work from a fresh Google Colab runtime using **Runtime → Restart and run all**.

It clones the repo if needed, enters the repo root, installs dependencies, loads `paper.yaml`, and generates:

```text
reports/01_AGI_vs_ASI.md
```

## 1. Clone repo, enter repo root, and install dependencies

````python
from pathlib import Path
import os
import subprocess
import sys

REPO = "agi-vs-asi"
GITHUB_USER = "thinkthoughts"
BRANCH = "main"

CONTENT = Path("/content")
ROOT = CONTENT / REPO

if not ROOT.exists():
    url = f"https://github.com/{GITHUB_USER}/{REPO}.git"
    print(f"Cloning {url}")
    subprocess.run(["git", "clone", "--branch", BRANCH, url, str(ROOT)], check=True)
else:
    print(f"Repo already exists: {ROOT}")

os.chdir(ROOT)

subprocess.run(
    [sys.executable, "-m", "pip", "install", "-q", "pyyaml", "pandas", "matplotlib"],
    check=True
)

print("Working directory:", Path.cwd())
print("Repo contents:")
for item in sorted(Path.cwd().iterdir()):
    print("-", item.name)
````

## 2. Load `paper.yaml`

````python
import yaml

PAPER_YAML = ROOT / "paper.yaml"

if not PAPER_YAML.exists():
    raise FileNotFoundError(
        f"paper.yaml not found at {PAPER_YAML}. "
        "Make sure the repo contains paper.yaml at the repo root."
    )

with open(PAPER_YAML, "r", encoding="utf-8") as f:
    data = yaml.safe_load(f)

print("Loaded sections:")
for key in data.keys():
    print("-", key)
````

## 3. Paper and repo frame

````python
paper_id = data.get("id", "2606-12683")
title = data.get("title", "Constraints Demystify Superintelligence")
paper = data.get("paper", "arXiv:2606.12683")

links = data.get("links", {})
arxiv_url = links.get("arxiv", "https://arxiv.org/abs/2606.12683")
pdf_url = links.get("pdf", "https://arxiv.org/pdf/2606.12683")

claim = data.get(
    "claim",
    "The transition from artificial general intelligence (AGI) to artificial superintelligence (ASI) "
    "depends not only on increasing capability, but also on the constraints that shape scaling, "
    "recursive improvement, and coordination."
)

statement = data.get(
    "statement",
    "Superintelligence is shaped by both growth and constraint. "
    "Understanding bottlenecks is necessary for understanding future capability."
)

print("ID:", paper_id)
print("Title:", title)
print("Paper:", paper)
print("arXiv:", arxiv_url)
print("PDF:", pdf_url)
print()
print("Claim:")
print(claim)
print()
print("Engineering statement:")
print(statement)
````

## 4. Working definitions

The paper uses AGI and ASI informally, but carefully.

For this repo, the important move is to avoid treating “AGI” and “ASI” as the same thing.

```text
AGI = roughly human-level general intelligence

ASI = far beyond human-level AGI across broad domains,
      including capabilities beyond large human-expert collectives

Universal AI = theoretical upper limit, not a practical engineering system
```

````python
import pandas as pd

vocabulary = pd.DataFrame(
    [
        {
            "Term": "AGI",
            "Working definition": "Human-level artificial general intelligence; roughly as intelligent as a single human across cognitive tasks.",
            "Repo role": "Starting point of the transition.",
        },
        {
            "Term": "ASI",
            "Working definition": "Artificial general superintelligence; broadly beyond human-level AGI and large expert collectives.",
            "Repo role": "Potential destination or regime beyond AGI.",
        },
        {
            "Term": "Universal AI",
            "Working definition": "Theoretical limit of machine intelligence in the AIXI / Legg-Hutter framing.",
            "Repo role": "Upper-bound concept, not a direct engineering roadmap.",
        },
    ]
)

vocabulary
````

## 5. Capability hierarchy

This is a conceptual ladder, not a measured scale.

The purpose is to keep the repo oriented:

```text
Current frontier systems
    ↓
AGI
    ↓
ASI
    ↓
Universal AI
```

````python
import matplotlib.pyplot as plt

levels = [
    "Current frontier\nsystems",
    "AGI",
    "ASI",
    "Universal AI",
]

scores = [1, 2, 3, 4]

plt.figure(figsize=(8, 4))
plt.plot(scores, scores, marker="o")
plt.xticks(scores, levels)
plt.yticks([])
plt.title("Conceptual capability hierarchy")
plt.xlabel("Conceptual level")
plt.tight_layout()
plt.show()
````

## 6. AGI checklist

AGI is not a single narrow benchmark.

A useful AGI discussion should ask whether a system can generalize across broad cognitive tasks.

````python
agi_checklist = pd.DataFrame(
    [
        ["Reasoning", "Can the system solve unfamiliar problems?"],
        ["Learning", "Can the system acquire new competence from experience?"],
        ["Planning", "Can the system plan over long horizons?"],
        ["Tool use", "Can the system use external tools effectively?"],
        ["Adaptation", "Can the system perform across changing contexts?"],
        ["Communication", "Can the system coordinate with humans or other agents?"],
    ],
    columns=["Capability", "Question"],
)

agi_checklist
````

## 7. What makes ASI different?

ASI is not merely “AGI, but smarter.”

The paper emphasizes digital advantages that scale differently from biological intelligence:

- speed,
- memory,
- replication,
- bandwidth,
- parallelism,
- collective coordination.

Those differences can compound.

````python
comparison = pd.DataFrame(
    [
        ["Reasoning", "Human-level", "Beyond human-level across broad domains"],
        ["Speed", "Human biological timescale", "Potentially much faster internal processing"],
        ["Working memory", "Limited", "Potentially much larger"],
        ["Replication", "Biologically constrained", "Digital copying of code and state"],
        ["Parallelism", "Limited by human population and training", "Large numbers of instances can run in parallel"],
        ["Coordination", "Low-bandwidth human institutions", "Potentially high-bandwidth agent collectives"],
        ["Learning transfer", "Slow cultural transfer", "Potentially high-bandwidth sharing or replay"],
    ],
    columns=["Property", "AGI", "ASI"],
)

comparison
````

## 8. Universal AI as upper bound

Universal AI helps frame the theoretical endpoint of machine intelligence.

But it is not the same as practical ASI.

Engineering interpretation:

> Universal AI is an upper-bound concept, not a deployment plan.

The repo can use it to reason about limits while keeping practical focus on pathways and constraints.

````python
universal_ai_notes = [
    "AIXI / Universal AI is theoretically useful.",
    "It provides an idealized upper bound for general intelligence.",
    "It is incomputable in its full form.",
    "It helps distinguish practical ASI from theoretical optimality.",
    "It should not be confused with a near-term engineering architecture.",
]

for item in universal_ai_notes:
    print("-", item)
````

## 9. Bridge to Notebook 07

Notebook 07 should focus on digital intelligence advantages.

The key bridge is:

```text
AGI vs ASI is not only about better reasoning.

It is also about how digital systems scale:
speed, memory, copying, sharing, and parallelism.
```

````python
digital_advantages = pd.DataFrame(
    [
        ["Input / output speed", "More information can be processed and produced per unit time."],
        ["Internal processing speed", "Thinking or reasoning can run faster with more compute."],
        ["Working memory", "Large memory and high read/write bandwidth can exceed human limits."],
        ["Lossless replication", "Code and memory states can be copied."],
        ["High-bandwidth sharing", "Experiences, traces, and updates can be shared or replayed."],
        ["Parallel instances", "Many instances can work simultaneously."],
    ],
    columns=["Digital advantage", "Why it matters"],
)

digital_advantages
````

## 10. Engineering statement

AGI and ASI should be distinguished by both capability level and scaling structure.

The repo's working statement:

```text
Capability growth does not erase constraints.
Understanding AGI and ASI requires understanding both growth and friction.
```

````python
repo_statement = (
    "Capability growth does not erase constraints. "
    "Understanding AGI and ASI requires understanding both growth and friction."
)

print(repo_statement)
````

## 11. Development questions

````python
questions = [
    "Can AGI exist without quickly leading to ASI?",
    "Can ASI emerge from coordinated collectives even if no single model is individually ASI-level?",
    "Is Universal AI useful as a practical guide, or mainly as a theoretical bound?",
    "Which digital advantage matters most: speed, memory, replication, sharing, or coordination?",
    "What would count as evidence that AGI-to-ASI transition has begun?",
]

for i, question in enumerate(questions, start=1):
    print(f"{i}. {question}")
````

## 12. Write `reports/01_AGI_vs_ASI.md`

````python
reports_dir = ROOT / "reports"
reports_dir.mkdir(exist_ok=True)

lines = [
    "# 01 — AGI vs ASI",
    "",
    f"**Paper:** {paper}",
    f"**Repo title:** {title}",
    f"**arXiv:** {arxiv_url}",
    "",
    "## Core distinction",
    "",
    "- **AGI:** human-level artificial general intelligence.",
    "- **ASI:** artificial general superintelligence beyond human-level AGI and large expert collectives.",
    "- **Universal AI:** theoretical upper-bound concept, not a practical engineering roadmap.",
    "",
    "## Core claim",
    "",
    claim,
    "",
    "## Working statement",
    "",
    repo_statement,
    "",
    "## Digital advantages that motivate the next notebook",
    "",
]

for _, row in digital_advantages.iterrows():
    lines.append(f"- **{row['Digital advantage']}:** {row['Why it matters']}")

lines.extend([
    "",
    "## Development questions",
    "",
])

for i, question in enumerate(questions, start=1):
    lines.append(f"{i}. {question}")

out = reports_dir / "01_AGI_vs_ASI.md"
out.write_text("\n".join(lines), encoding="utf-8")

print("Wrote:", out)
````
