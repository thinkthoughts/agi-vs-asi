# 01 — AGI vs ASI

Clarifying the distinction between:

- AGI
- ASI
- Universal AI

before analyzing scaling, recursive improvement, and collective intelligence.

This notebook is designed to run from a fresh Google Colab runtime.

## 1. Clone repo, enter repo root, and install dependencies

```python
from pathlib import Path
import os
import subprocess
import sys

REPO = "agi-vs-asi"
GITHUB_USER = "thinkthoughts"
BRANCH = "main"

CONTENT = Path("/content")
ROOT = CONTENT / REPO

if not ROOT.exists():
    url = f"https://github.com/{GITHUB_USER}/{REPO}.git"
    print(f"Cloning {url}")
    subprocess.run(
        ["git", "clone", "--branch", BRANCH, url, str(ROOT)],
        check=True,
    )
else:
    print(f"Repo already exists: {ROOT}")

os.chdir(ROOT)

subprocess.run(
    [sys.executable, "-m", "pip", "install", "-q", "pyyaml", "pandas"],
    check=True,
)

print("Working directory:", Path.cwd())
```

## 2. Load `paper.yaml`

```python
import yaml

PAPER_YAML = Path("paper.yaml")

if not PAPER_YAML.exists():
    raise FileNotFoundError(
        "paper.yaml not found in the repo root. "
        "Confirm that the repository has paper.yaml at /content/agi-vs-asi/paper.yaml."
    )

with open(PAPER_YAML, "r", encoding="utf-8") as f:
    data = yaml.safe_load(f)

print(data.get("title", "No title found"))
```

## 3. Paper summary

```python
print("Paper:", data.get("paper", "arXiv:2606.12683"))
print()

print("Claim:")
print(data.get("claim", "No claim found in paper.yaml."))
print()

print("Statement:")
print(data.get("statement", "No statement found in paper.yaml."))
```

## 4. AGI / ASI vocabulary

The paper is titled **From AGI to ASI**, so this notebook first separates the terms.

The goal is not to settle every definition. The goal is to give the repo a stable working vocabulary.

```python
import pandas as pd

vocabulary = pd.DataFrame(
    [
        [
            "AGI",
            "Human-level general intelligence",
            "Starting point",
        ],
        [
            "ASI",
            "Intelligence beyond large human collectives",
            "Growth destination",
        ],
        [
            "Universal AI",
            "Theoretical optimum agent",
            "Upper-bound concept",
        ],
    ],
    columns=["Term", "Definition", "Role"],
)

vocabulary
```

## 5. AGI vs ASI comparison

ASI is not only “AGI but smarter.”

The distinction also involves scale, speed, replication, memory, coordination, and parallelism.

```python
comparison = pd.DataFrame(
    [
        ["Reasoning", "Human-level", "Beyond human-level"],
        ["Speed", "Human timescale", "Potentially much faster"],
        ["Memory", "Limited", "Expandable"],
        ["Replication", "No", "Digital copies possible"],
        ["Parallelism", "Limited", "Massive"],
        ["Coordination", "Human institutions", "Large agent collectives"],
    ],
    columns=["Property", "AGI", "ASI"],
)

comparison
```

## 6. Digital advantages

This table bridges to Notebook 07.

The central question is how digital intelligence scales differently from biological intelligence.

```python
digital_advantages = pd.DataFrame(
    [
        ["Speed", "More computation per unit time"],
        ["Memory", "Larger working memory"],
        ["Replication", "Copyable systems"],
        ["Parallelism", "Many simultaneous instances"],
        ["Sharing", "High-bandwidth transfer"],
    ],
    columns=["Advantage", "Why it matters"],
)

digital_advantages
```

## 7. AGI-to-ASI pathway map

Notebook 01 establishes the vocabulary.

The next notebooks each focus on one pathway or constraint.

```python
pathways = pd.DataFrame(
    [
        ["Digital advantages", "07_digital_advantages.ipynb"],
        ["Scaling compute", "13_scaling_compute.ipynb"],
        ["Algorithmic progress", "17_algorithmic_progress.ipynb"],
        ["Recursive improvement", "23_recursive_improvement.ipynb"],
        ["Collectives", "29_multi_agent_collectives.ipynb"],
        ["Constraints", "31_constraints_and_bottlenecks.ipynb"],
    ],
    columns=["Pathway", "Notebook"],
)

pathways
```

## 8. Engineering statement

```python
statement = '''
Capability growth does not erase constraints.

Understanding AGI and ASI requires understanding
both growth and friction.
'''

print(statement)
```

## 9. Development questions

```python
questions = [
    "Can AGI exist without ASI?",
    "Can ASI emerge from collectives?",
    "Is Universal AI physically achievable?",
    "Which digital advantage matters most?",
]

for i, q in enumerate(questions, start=1):
    print(f"{i}. {q}")
```

## 10. Write report

```python
reports = Path("reports")
reports.mkdir(exist_ok=True)

out = reports / "01_AGI_vs_ASI.md"

lines = [
    "# 01 — AGI vs ASI",
    "",
    "## Working vocabulary",
    "",
]

for _, row in vocabulary.iterrows():
    lines.append(f"- **{row['Term']}:** {row['Definition']} ({row['Role']})")

lines.extend([
    "",
    "## AGI vs ASI",
    "",
])

for _, row in comparison.iterrows():
    lines.append(f"- **{row['Property']}:** AGI — {row['AGI']}; ASI — {row['ASI']}")

lines.extend([
    "",
    "## Digital advantages",
    "",
])

for _, row in digital_advantages.iterrows():
    lines.append(f"- **{row['Advantage']}:** {row['Why it matters']}")

lines.extend([
    "",
    "## Pathway map",
    "",
])

for _, row in pathways.iterrows():
    lines.append(f"- **{row['Pathway']}:** `{row['Notebook']}`")

lines.extend([
    "",
    "## Engineering statement",
    "",
    statement.strip(),
    "",
    "## Development questions",
    "",
])

for i, q in enumerate(questions, start=1):
    lines.append(f"{i}. {q}")

out.write_text("\n".join(lines), encoding="utf-8")

print("Wrote", out)
```
