## Vocabulary

```python
import pandas as pd

vocab = pd.DataFrame([
    ['AGI','Human-level general intelligence'],
    ['ASI','Intelligence beyond large human collectives'],
    ['Universal AI','Theoretical optimal agent']
], columns=['Term','Definition'])

print(vocab.to_string(index=False))
```

## Capability hierarchy

```python
import matplotlib.pyplot as plt

levels = ['Current Frontier','AGI','ASI','Universal AI']
plt.figure(figsize=(6,3))
plt.plot(range(len(levels)), range(len(levels)), marker='o')
plt.xticks(range(len(levels)), levels, rotation=20)
plt.yticks([])
plt.title('Capability hierarchy')
plt.tight_layout()
plt.show()
```
