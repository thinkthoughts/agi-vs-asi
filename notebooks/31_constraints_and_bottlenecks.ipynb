## 1. Clone repo, enter repo root, and install dependencies

````python
from pathlib import Path
import os
import subprocess
import sys

REPO = "agi-vs-asi"
GITHUB_USER = "thinkthoughts"
BRANCH = "main"

CONTENT = Path("/content")
ROOT = CONTENT / REPO

if not ROOT.exists():
    url = f"https://github.com/{GITHUB_USER}/{REPO}.git"
    print(f"Cloning {url}")
    subprocess.run(
        ["git", "clone", "--branch", BRANCH, url, str(ROOT)],
        check=True,
    )
else:
    print(f"Repo already exists: {ROOT}")

os.chdir(ROOT)

subprocess.run(
    [sys.executable, "-m", "pip", "install", "-q", "pyyaml", "pandas"],
    check=True,
)

print("Working directory:", Path.cwd())
````

## 2. Load `paper.yaml` and `statement.yaml`

````python
import yaml

PAPER_YAML = Path("paper.yaml")
STATEMENT_YAML = Path("statement.yaml")

if not PAPER_YAML.exists():
    raise FileNotFoundError(
        "paper.yaml not found in the repo root. "
        "Confirm that the repository has paper.yaml at /content/agi-vs-asi/paper.yaml."
    )

if not STATEMENT_YAML.exists():
    raise FileNotFoundError(
        "statement.yaml not found in the repo root. "
        "Confirm that the repository has statement.yaml at /content/agi-vs-asi/statement.yaml."
    )

with open(PAPER_YAML, "r", encoding="utf-8") as f:
    paper = yaml.safe_load(f)

with open(STATEMENT_YAML, "r", encoding="utf-8") as f:
    statement = yaml.safe_load(f)

print(paper.get("title", "No title found"))
print(paper.get("paper", "arXiv:2606.12683"))
print()
print(statement.get("equation", "capability = growth - friction"))
````

## 5. Growth pathway vs constraint

Each growth pathway has different bottlenecks.

This table connects the previous notebooks into one map.

````python
pathway_constraints = pd.DataFrame(
    [
        [
            "Scaling compute",
            "Energy, hardware, capital, data",
            "Scaling can stall if physical and economic inputs cannot keep growing.",
        ],
        [
            "Algorithmic progress",
            "Research difficulty, evaluation, benchmark saturation",
            "Efficiency gains depend on finding better methods and measuring them honestly.",
        ],
        [
            "Feedback lifts",
            "Verification, quality control, security, metric gaming",
            "Feedback can amplify capability, but it can also amplify errors.",
        ],
        [
            "Multi-agent collectives",
            "Coordination, communication, integration, incentives",
            "More agents help only when coordination produces more value than it costs.",
        ],
        [
            "Forecasting progress",
            "Measurement uncertainty, changing baselines, hidden variables",
            "Forecasts depend on tracking quantities that are difficult to observe directly.",
        ],
    ],
    columns=["Growth pathway", "Dominant constraints", "Why it matters"],
)

pathway_constraints
````

## 6. Constraint interaction map

Constraints are coupled.

A hardware bottleneck can become an energy bottleneck.

A data bottleneck can become an evaluation bottleneck.

A coordination bottleneck can become a forecasting bottleneck.

````python
constraint_interactions = pd.DataFrame(
    [
        ["Energy", "Hardware", "Power and cooling constrain how much hardware can be deployed."],
        ["Hardware", "Compute", "Chips, memory, and interconnect constrain training and inference scale."],
        ["Data", "Algorithms", "Data quality changes what algorithms can learn."],
        ["Evaluation", "Feedback lifts", "Bad metrics can turn feedback into optimization of the wrong target."],
        ["Coordination", "Collectives", "Communication and integration determine whether groups outperform individuals."],
        ["Economics", "Scaling", "Capital costs determine what scale is actually deployable."],
        ["Physics", "Hardware and energy", "Thermodynamics and information propagation impose hard limits."],
        ["Complexity", "Algorithms", "Some tasks resist brute-force scaling or simple approximation."],
    ],
    columns=["Constraint", "Amplifies or limits", "Connection"],
)

constraint_interactions
````

## 8. Physical vs organizational constraints

## 9. Constraint tradeoffs

A common mistake is to imagine a bottleneck disappears.

Often it moves.

Example:

```text
better algorithms
    ↓
less training compute needed
    ↓
more demand for inference, evaluation, and deployment
```

````python
tradeoffs = pd.DataFrame(
    [
        [
            "Algorithmic efficiency reduces training cost",
            "Inference demand may grow because more applications become economical.",
        ],
        [
            "Synthetic data reduces data scarcity",
            "Evaluation and contamination risks increase.",
        ],
        [
            "More agents increase parallelism",
            "Coordination overhead grows.",
        ],
        [
            "Better hardware increases compute stock",
            "Energy, cooling, and capital costs become more salient.",
        ],
        [
            "Better evaluations improve optimization",
            "Systems may overfit the new evaluations.",
        ],
    ],
    columns=["Apparent solution", "Possible shifted bottleneck"],
)

tradeoffs
````

## 12. Write report

````python
reports = Path("reports")
reports.mkdir(exist_ok=True)

out = reports / "31_constraints_and_bottlenecks.md"

lines = [
    "# 31 — Constraints and Bottlenecks",
    "",
    "This notebook maps what limits capability growth across the repo's main pathways.",
    "",
    "## Constraint categories",
    "",
]

for _, row in constraint_categories.iterrows():
    lines.append(f"- **{row['Constraint']}:** {row['Meaning']}")

lines.extend([
    "",
    "## Pathway constraints",
    "",
])

for _, row in pathway_constraints.iterrows():
    lines.append(
        f"- **{row['Growth pathway']}:** {row['Dominant constraints']} — {row['Why it matters']}"
    )

lines.extend([
    "",
    "## Growth vs friction",
    "",
])

for _, row in growth_vs_friction.iterrows():
    lines.append(
        f"- **{row['Growth source']}:** growth — {row['Growth effect']}; friction — {row['Friction']}."
    )

lines.extend([
    "",
    "## Constraint tradeoffs",
    "",
])

for _, row in tradeoffs.iterrows():
    lines.append(
        f"- **{row['Apparent solution']}:** shifted bottleneck — {row['Possible shifted bottleneck']}"
    )

lines.extend([
    "",
    "## Development questions",
    "",
])

for i, question in enumerate(questions, start=1):
    lines.append(f"{i}. {question}")

lines.extend([
    "",
    "## Working statement",
    "",
    "Capability growth is constrained by coupled physical, economic, technical, and organizational bottlenecks.",
])

out.write_text("\n".join(lines), encoding="utf-8")

print("Wrote", out)
````
