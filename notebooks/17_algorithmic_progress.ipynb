# 17 — Algorithmic Progress

This notebook separates raw compute growth from algorithmic progress.

Notebook 13 focused on scaling compute.

Notebook 17 asks:

> What happens when systems get more capable per unit of compute?

The focus is on:

- algorithmic efficiency,
- paradigm evolution,
- test-time scaling,
- continual learning,
- agent scaffolding,
- world models,
- long-context and memory systems.

This notebook is designed to run from a fresh Google Colab runtime.

## 1. Clone repo, enter repo root, and install dependencies

````python
from pathlib import Path
import os
import subprocess
import sys

REPO = "agi-vs-asi"
GITHUB_USER = "thinkthoughts"
BRANCH = "main"

CONTENT = Path("/content")
ROOT = CONTENT / REPO

if not ROOT.exists():
    url = f"https://github.com/{GITHUB_USER}/{REPO}.git"
    print(f"Cloning {url}")
    subprocess.run(
        ["git", "clone", "--branch", BRANCH, url, str(ROOT)],
        check=True,
    )
else:
    print(f"Repo already exists: {ROOT}")

os.chdir(ROOT)

subprocess.run(
    [sys.executable, "-m", "pip", "install", "-q", "pyyaml", "pandas"],
    check=True,
)

print("Working directory:", Path.cwd())
````

## 2. Load `paper.yaml`

````python
import yaml

PAPER_YAML = Path("paper.yaml")

if not PAPER_YAML.exists():
    raise FileNotFoundError(
        "paper.yaml not found in the repo root. "
        "Confirm that the repository has paper.yaml at /content/agi-vs-asi/paper.yaml."
    )

with open(PAPER_YAML, "r", encoding="utf-8") as f:
    data = yaml.safe_load(f)

print(data.get("title", "No title found"))
print(data.get("paper", "arXiv:2606.12683"))
````

## 3. Core question

Scaling compute asks:

```text
What happens when more compute is supplied?
```

Algorithmic progress asks:

```text
What happens when the same compute produces more capability?
```

In the paper's framing, algorithmic efficiency can act like extra hardware.

That makes algorithmic progress a multiplier on the AGI-to-ASI transition.

````python
core_question = (
    "How do algorithmic efficiency and paradigm evolution change the relationship "
    "between compute and capability?"
)

print(core_question)
````

## 4. Algorithmic progress categories

This table defines the categories this repo should track.

````python
import pandas as pd

algorithmic_categories = pd.DataFrame(
    [
        [
            "Algorithmic efficiency",
            "Less compute needed to reach the same capability",
            "Acts like effective hardware growth",
        ],
        [
            "Architecture evolution",
            "Model structures that change scaling behavior",
            "Can alter memory, context, efficiency, and reasoning",
        ],
        [
            "Optimization improvements",
            "Better training methods, objectives, or data mixtures",
            "Can improve learning per unit compute",
        ],
        [
            "Test-time scaling",
            "More search, reasoning, sampling, planning, and tool use after training",
            "Can improve task performance without retraining",
        ],
        [
            "Continual learning",
            "Learning from new experience without full retraining",
            "Can turn deployment experience into capability growth",
        ],
        [
            "Agent scaffolding",
            "Tool use, planning loops, memory, decomposition, and orchestration",
            "Can make a base model act like a more capable system",
        ],
        [
            "World models",
            "Internal or external models of environment dynamics",
            "Can support planning, simulation, and robust decision-making",
        ],
    ],
    columns=["Category", "Meaning", "Why it matters"],
)

algorithmic_categories
````

## 5. Current paradigm vs possible evolution

The paper distinguishes between evolution of the current paradigm and sharper paradigm shifts.

For this repo, the useful split is:

```text
current paradigm evolution
vs
paradigm shift
```

````python
paradigm_table = pd.DataFrame(
    [
        [
            "Current paradigm",
            "Large pretrained models, supervised prediction, RL tuning, tool use, retrieval, test-time reasoning",
            "Still being extended rapidly",
        ],
        [
            "Paradigm evolution",
            "Longer context, memory, agents, continual learning, better inference-time search",
            "Likely near-term and directly measurable",
        ],
        [
            "Paradigm shift",
            "Substantially different architectures, training regimes, hardware substrates, or agency models",
            "Harder to forecast but potentially high-impact",
        ],
    ],
    columns=["Layer", "Description", "Forecasting status"],
)

paradigm_table
````

## 6. Growth vs friction

Algorithmic progress is powerful because it can multiply compute.

But each algorithmic path has frictions.

````python
algorithmic_friction = pd.DataFrame(
    [
        [
            "Algorithmic efficiency",
            "More capability per unit compute",
            "diminishing returns, measurement uncertainty, benchmark saturation",
        ],
        [
            "Long context / memory",
            "More information available during reasoning",
            "retrieval errors, stale memory, context dilution, evaluation gaps",
        ],
        [
            "Test-time scaling",
            "More search and planning per task",
            "latency, cost, reliability, hallucinated reasoning traces",
        ],
        [
            "Continual learning",
            "Deployment experience becomes learning signal",
            "catastrophic forgetting, stale updates, evaluation drift",
        ],
        [
            "Agent scaffolding",
            "Models become more capable through tools and planning loops",
            "fragility, compounding errors, tool misuse, security risk",
        ],
        [
            "World models",
            "Better planning through simulated futures",
            "model misspecification, causal errors, distribution shift",
        ],
    ],
    columns=["Path", "Growth effect", "Friction"],
)

algorithmic_friction
````

## 7. Why algorithmic progress matters for AGI → ASI

If compute growth slows, algorithmic progress may still improve capability.

If compute growth continues, algorithmic progress compounds it.

That is why the paper treats algorithmic development as a separate pathway rather than a footnote to scaling.

````python
interaction_table = pd.DataFrame(
    [
        [
            "Compute grows, algorithms stagnate",
            "Capability may improve, but returns may slow",
        ],
        [
            "Compute stagnates, algorithms improve",
            "Capability may still grow through efficiency gains",
        ],
        [
            "Compute grows and algorithms improve",
            "Effective compute compounds fastest",
        ],
        [
            "Algorithms create recursive improvement",
            "AI progress may accelerate through feedback loops",
        ],
    ],
    columns=["Scenario", "Implication"],
)

interaction_table
````

## 8. Bridge to recursive improvement

Algorithmic progress becomes recursive improvement when AI systems contribute to improving future AI systems.

That bridge matters:

```text
algorithmic progress
        ↓
AI-assisted AI research
        ↓
recursive improvement
        ↓
possible acceleration
```

Notebook 23 will focus on that feedback loop.

````python
recursive_bridge = pd.DataFrame(
    [
        ["Code", "AI helps design architectures, training code, optimizers, evaluations."],
        ["Data", "AI helps generate, curate, filter, or simulate training data."],
        ["Hardware", "AI helps design chips, datacenters, or energy systems."],
        ["Agents", "AI helps organize division of labor among many systems."],
        ["Evaluation", "AI helps build better benchmarks, audits, and measurement systems."],
    ],
    columns=["Recursive channel", "How algorithmic progress becomes feedback"],
)

recursive_bridge
````

## 9. Development questions

The repo should avoid treating algorithmic progress as magic.

The useful questions are specific.

````python
questions = [
    "Which algorithmic improvements are equivalent to effective compute growth?",
    "How should algorithmic efficiency be measured across changing benchmarks?",
    "Does test-time scaling improve reliability or mainly increase cost and latency?",
    "Can continual learning improve deployed systems without making them unstable?",
    "Which agent scaffolds create genuine capability rather than brittle demos?",
    "What would count as evidence for a paradigm shift rather than normal progress?",
]

for i, question in enumerate(questions, start=1):
    print(f"{i}. {question}")
````

## 10. Write report

````python
reports = Path("reports")
reports.mkdir(exist_ok=True)

out = reports / "17_algorithmic_progress.md"

lines = [
    "# 17 — Algorithmic Progress",
    "",
    "Algorithmic progress changes the relationship between compute and capability.",
    "",
    "It can act like effective hardware growth by producing more capability per unit compute.",
    "",
    "## Categories",
    "",
]

for _, row in algorithmic_categories.iterrows():
    lines.append(f"- **{row['Category']}:** {row['Meaning']} — {row['Why it matters']}")

lines.extend([
    "",
    "## Growth vs friction",
    "",
])

for _, row in algorithmic_friction.iterrows():
    lines.append(
        f"- **{row['Path']}:** growth — {row['Growth effect']}; friction — {row['Friction']}."
    )

lines.extend([
    "",
    "## Recursive bridge",
    "",
])

for _, row in recursive_bridge.iterrows():
    lines.append(f"- **{row['Recursive channel']}:** {row['How algorithmic progress becomes feedback']}")

lines.extend([
    "",
    "## Development questions",
    "",
])

for i, question in enumerate(questions, start=1):
    lines.append(f"{i}. {question}")

lines.extend([
    "",
    "## Working statement",
    "",
    "Algorithmic progress is not separate from scaling: it changes what scaling means by increasing capability per unit compute.",
])

out.write_text("\n".join(lines), encoding="utf-8")

print("Wrote", out)
````
