# 23 — Feedback Lifts

This notebook replaces the more speculative phrase “recursive improvement” with a more concrete engineering question:

> Which feedback lifts capability?

A feedback lift is a mechanism where present AI capability helps improve future AI capability.

Examples:

- code,
- data,
- evaluation,
- hardware,
- coordination.

This notebook is designed to run from a fresh Google Colab runtime.

## 1. Clone repo, enter repo root, and install dependencies

````python
from pathlib import Path
import os
import subprocess
import sys

REPO = "agi-vs-asi"
GITHUB_USER = "thinkthoughts"
BRANCH = "main"

CONTENT = Path("/content")
ROOT = CONTENT / REPO

if not ROOT.exists():
    url = f"https://github.com/{GITHUB_USER}/{REPO}.git"
    print(f"Cloning {url}")
    subprocess.run(
        ["git", "clone", "--branch", BRANCH, url, str(ROOT)],
        check=True,
    )
else:
    print(f"Repo already exists: {ROOT}")

os.chdir(ROOT)

subprocess.run(
    [sys.executable, "-m", "pip", "install", "-q", "pyyaml", "pandas"],
    check=True,
)

print("Working directory:", Path.cwd())
````

## 2. Load `paper.yaml` and `statement.yaml`

````python
import yaml

PAPER_YAML = Path("paper.yaml")
STATEMENT_YAML = Path("statement.yaml")

if not PAPER_YAML.exists():
    raise FileNotFoundError(
        "paper.yaml not found in the repo root. "
        "Confirm that the repository has paper.yaml at /content/agi-vs-asi/paper.yaml."
    )

if not STATEMENT_YAML.exists():
    raise FileNotFoundError(
        "statement.yaml not found in the repo root. "
        "Confirm that the repository has statement.yaml at /content/agi-vs-asi/statement.yaml."
    )

with open(PAPER_YAML, "r", encoding="utf-8") as f:
    paper = yaml.safe_load(f)

with open(STATEMENT_YAML, "r", encoding="utf-8") as f:
    statement = yaml.safe_load(f)

print(paper.get("title", "No title found"))
print(paper.get("paper", "arXiv:2606.12683"))
print()
print(statement.get("statement", "No statement found."))
````

## 4. Feedback lift categories

This is the primary artifact of Notebook 23.

````python
import pandas as pd

feedback_lifts = pd.DataFrame(
    [
        [
            "Code",
            "AI assists software creation and system design",
            "training pipelines, model code, agent scaffolds, infrastructure scripts",
        ],
        [
            "Data",
            "AI generates, curates, filters, or simulates learning signal",
            "synthetic datasets, search traces, environment interactions, data cleaning",
        ],
        [
            "Evaluation",
            "AI improves testing, benchmarking, auditing, and failure discovery",
            "generated tests, red-team prompts, benchmark construction, anomaly detection",
        ],
        [
            "Hardware",
            "AI assists chip, datacenter, cooling, energy, or systems design",
            "accelerator layouts, scheduling, thermal design, energy optimization",
        ],
        [
            "Coordination",
            "AI organizes tasks, agents, teams, and workflows",
            "multi-agent planning, task routing, division of labor, project management",
        ],
    ],
    columns=["Feedback lift", "Mechanism", "Examples"],
)

feedback_lifts
````

## 5. Feedback lift chain

A lift matters when present capability improves the next capability-building step.

````python
lift_chain = pd.DataFrame(
    [
        ["Code", "Faster development", "Better systems and shorter iteration cycles"],
        ["Data", "More or better learning signal", "Better training, tuning, or evaluation"],
        ["Evaluation", "Better measurement", "Better optimization and safer deployment"],
        ["Hardware", "More effective compute", "Larger or cheaper capability growth"],
        ["Coordination", "More organized collective work", "Higher collective capability"],
    ],
    columns=["Feedback lift", "Immediate effect", "Future effect"],
)

lift_chain
````

## 6. Growth vs friction

The repo equation remains:

```text
capability = growth - friction
```

Feedback lifts create growth.

They also introduce frictions.

````python
growth_friction = pd.DataFrame(
    [
        ["Code", "Faster iteration and system improvement", "verification, reliability, security, maintainability"],
        ["Data", "More learning signal and coverage", "quality, contamination, distribution shift, evaluation leakage"],
        ["Evaluation", "Better measurement and optimization targets", "metric gaming, benchmark overfitting, false confidence"],
        ["Hardware", "More compute and efficiency", "fabrication, supply chains, energy, cooling, capital cost"],
        ["Coordination", "More collective output and specialization", "communication overhead, incentives, duplicated work"],
    ],
    columns=["Feedback lift", "Growth effect", "Friction"],
)

growth_friction
````

## 7. Evidence thresholds

This notebook avoids treating feedback lifts as automatic.

A meaningful feedback lift should leave evidence.

````python
evidence_thresholds = pd.DataFrame(
    [
        ["Code", "AI-written or AI-designed improvements are adopted into production systems and improve measured outcomes."],
        ["Data", "AI-generated or AI-curated data improves generalization or capability on held-out evaluations."],
        ["Evaluation", "AI-generated tests reveal failures humans missed and lead to measurable system improvements."],
        ["Hardware", "AI-assisted hardware or infrastructure designs outperform strong human baselines."],
        ["Coordination", "AI-managed teams or agent collectives outperform isolated agents or human-managed workflows."],
    ],
    columns=["Feedback lift", "Evidence threshold"],
)

evidence_thresholds
````

## 8. Ordinary assistance vs feedback lift

Not every use of AI in AI work is a feedback lift.

A feedback lift requires a measurable improvement in future capability-building capacity.

````python
ordinary_vs_lift = pd.DataFrame(
    [
        [
            "AI helps write boilerplate",
            "Ordinary assistance",
            "Useful, but may not measurably improve future AI capability.",
        ],
        [
            "AI discovers a training bug that improves model performance",
            "Feedback lift",
            "Improves the capability-building process.",
        ],
        [
            "AI generates many synthetic examples",
            "Depends",
            "A lift only if the data improves robust held-out performance.",
        ],
        [
            "AI runs project management for many agents",
            "Depends",
            "A lift only if coordination improves output beyond the baseline.",
        ],
        [
            "AI designs a better benchmark",
            "Feedback lift",
            "Can improve measurement and optimization.",
        ],
    ],
    columns=["Case", "Classification", "Reason"],
)

ordinary_vs_lift
````

## 9. Bridge to multi-agent collectives

Notebook 29 should extend feedback lifts into collective systems.

```text
single-system feedback
        ↓
multi-agent feedback
        ↓
collective capability
```

Coordination is where feedback lifts and multi-agent collectives meet.

````python
bridge = pd.DataFrame(
    [
        ["Single system", "One model or agent improves code, data, or evaluation."],
        ["Workflow", "Multiple tools and agents improve a research or engineering process."],
        ["Collective", "Many agents coordinate across tasks, memory, evaluation, and specialization."],
        ["Institution", "Capability growth becomes embedded in organizations and infrastructure."],
    ],
    columns=["Level", "Description"],
)

bridge
````

## 11. Write report

````python
reports = Path("reports")
reports.mkdir(exist_ok=True)

out = reports / "23_feedback_lifts.md"

lines = [
    "# 23 — Feedback Lifts",
    "",
    "A feedback lift is a mechanism where present AI capability helps improve future AI capability.",
    "",
    "The repo uses this term instead of the more speculative phrase “recursive improvement.”",
    "",
    "## Feedback lift categories",
    "",
]

for _, row in feedback_lifts.iterrows():
    lines.append(f"- **{row['Feedback lift']}:** {row['Mechanism']} ({row['Examples']}).")

lines.extend([
    "",
    "## Growth vs friction",
    "",
])

for _, row in growth_friction.iterrows():
    lines.append(
        f"- **{row['Feedback lift']}:** growth — {row['Growth effect']}; friction — {row['Friction']}."
    )

lines.extend([
    "",
    "## Evidence thresholds",
    "",
])

for _, row in evidence_thresholds.iterrows():
    lines.append(f"- **{row['Feedback lift']}:** {row['Evidence threshold']}")

lines.extend([
    "",
    "## Working statement",
    "",
    "Capability growth can be amplified by feedback lifts that improve future capability-building capacity.",
    "",
    "## Development questions",
    "",
])

for i, question in enumerate(questions, start=1):
    lines.append(f"{i}. {question}")

out.write_text("\n".join(lines), encoding="utf-8")

print("Wrote", out)
````
