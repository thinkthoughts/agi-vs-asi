# 00 — Context

Fresh-runtime notebook for `agi_vs_asi`.

This notebook initializes the repo around arXiv:2606.12683, **From AGI to ASI**.

It is designed to work from a fresh Google Colab runtime using **Runtime → Restart and run all**.

It will:

1. clone the GitHub repo if needed,
2. install `pyyaml`,
3. enter the repo root,
4. load `paper.yaml`,
5. summarize the paper context,
6. map the notebook roadmap,
7. generate `reports/context_summary.md`.

## 1. Clone repo, enter repo root, and install dependencies

````python
from pathlib import Path
import os
import subprocess
import sys

REPO = "agi_vs_asi"
GITHUB_USER = "thinkthoughts"
BRANCH = "main"

CONTENT = Path("/content")
ROOT = CONTENT / REPO

if not ROOT.exists():
    url = f"https://github.com/{GITHUB_USER}/{REPO}.git"
    print(f"Cloning {url}")
    subprocess.run(["git", "clone", "--branch", BRANCH, url, str(ROOT)], check=True)
else:
    print(f"Repo already exists: {ROOT}")

os.chdir(ROOT)

subprocess.run([sys.executable, "-m", "pip", "install", "-q", "pyyaml"], check=True)

print("Working directory:", Path.cwd())
print("Repo contents:")
for item in sorted(Path.cwd().iterdir()):
    print("-", item.name)
````

## 2. Load `paper.yaml`

````python
import yaml

PAPER_YAML = ROOT / "paper.yaml"

if not PAPER_YAML.exists():
    raise FileNotFoundError(
        f"paper.yaml not found at {PAPER_YAML}. "
        "Make sure the repo contains paper.yaml at the repo root."
    )

with open(PAPER_YAML, "r", encoding="utf-8") as f:
    data = yaml.safe_load(f)

print("Loaded sections:")
for key in data.keys():
    print("-", key)
````

## 3. Paper metadata

````python
paper_id = data.get("id", "2606-12683")
title = data.get("title", "Constraints Demystify Superintelligence")
paper = data.get("paper", "arXiv:2606.12683")

links = data.get("links", {})
arxiv_url = links.get("arxiv", "https://arxiv.org/abs/2606.12683")
pdf_url = links.get("pdf", "https://arxiv.org/pdf/2606.12683")

print("ID:", paper_id)
print("Title:", title)
print("Paper:", paper)
print("arXiv:", arxiv_url)
print("PDF:", pdf_url)
````

## 4. Core frame

The paper asks how AI development might continue after human-level AGI.

The repo frame is:

```text
AGI
  ↓
growth pathways
  ↓
frictions and bottlenecks
  ↓
possible ASI trajectories
```

The engineering statement is that superintelligence is shaped by both growth and constraint.

````python
claim = data.get(
    "claim",
    "The transition from AGI to ASI depends not only on increasing capability, "
    "but also on the constraints that shape scaling, recursive improvement, and coordination."
)

statement = data.get(
    "statement",
    "Superintelligence is shaped by both growth and constraint. "
    "Understanding bottlenecks is necessary for understanding future capability."
)

equation = data.get("equation", "capability = growth - friction")

print("Claim:")
print(claim)
print()
print("Engineering statement:")
print(statement)
print()
print("Equation:")
print(equation)
````

## 5. Key vocabulary

````python
vocabulary = {
    "AGI": "human-level artificial general intelligence",
    "ASI": "artificial general superintelligence beyond large human-expert collectives",
    "Universal AI": "theoretical limit of machine intelligence in the AIXI / Legg-Hutter frame",
    "effective compute": "hardware, investment, and algorithmic efficiency combined as usable capability growth",
    "recursive improvement": "AI improving AI research, hardware, data, or agent organization",
    "multi-agent collectives": "capability emerging from coordinated groups of AI agents",
    "friction": "a bottleneck or constraint that slows or reshapes capability growth",
}

for key, value in vocabulary.items():
    print(f"{key}: {value}")
````

## 6. Growth pathways

````python
growth_pathways = [
    "scaling compute, models, and data",
    "algorithmic paradigm shifts and evolutions",
    "recursive self-improvement",
    "ASI via multi-agent collectives",
]

print("Pathways from AGI to ASI:")
for i, item in enumerate(growth_pathways, start=1):
    print(f"{i}. {item}")
````

## 7. Constraints and bottlenecks

````python
constraints = [
    "energy",
    "hardware",
    "data quality and availability",
    "algorithmic efficiency",
    "coordination",
    "communication",
    "physical limits",
    "complexity-theoretic limits",
    "forecast uncertainty",
]

print("Constraints to track:")
for item in constraints:
    print("-", item)
````

## 8. Notebook roadmap

````python
notebook_map = [
    ("00_context.ipynb", "Paper context, engineering statement, vocabulary, and roadmap."),
    ("01_AGI_vs_ASI.ipynb", "Clarify AGI, ASI, and Universal AI."),
    ("07_digital_advantages.ipynb", "Digital intelligence advantages: speed, memory, replication, sharing."),
    ("13_scaling_compute.ipynb", "Scaling compute, models, data, and effective compute."),
    ("17_algorithmic_progress.ipynb", "Algorithmic efficiency and paradigm shifts."),
    ("23_recursive_improvement.ipynb", "Recursive improvement through code, hardware, data, and specialization."),
    ("29_multi_agent_collectives.ipynb", "ASI as a collective or coordinated multi-agent phenomenon."),
    ("31_constraints_and_bottlenecks.ipynb", "Energy, hardware, data, coordination, physics, and complexity constraints."),
    ("37_forecasting_progress.ipynb", "Forecasting capability growth under uncertainty."),
    ("43_open_questions.ipynb", "Open research questions and future repo directions."),
]

for nb, purpose in notebook_map:
    print(f"{nb}: {purpose}")
````

## 9. Questions for development

````python
questions = [
    "Which bottleneck dominates current AGI-to-ASI forecasts: compute, energy, algorithms, data, or coordination?",
    "What would count as evidence for recursive self-improvement rather than ordinary tool-assisted AI research?",
    "Can multi-agent collectives produce ASI-level capability without each individual agent being ASI-level?",
    "How should effective compute be measured when algorithmic efficiency is changing rapidly?",
    "Which constraints are physical, which are economic, and which are governance-dependent?",
]

for i, question in enumerate(questions, start=1):
    print(f"{i}. {question}")
````

## 10. Write `reports/context_summary.md`

````python
reports_dir = ROOT / "reports"
reports_dir.mkdir(exist_ok=True)

lines = [
    "# 00 — Context",
    "",
    f"**Repo:** `{REPO}`",
    f"**Paper:** {paper}",
    f"**Title:** {title}",
    f"**arXiv:** {arxiv_url}",
    f"**PDF:** {pdf_url}",
    "",
    "## Core claim",
    "",
    claim,
    "",
    "## Engineering statement",
    "",
    statement,
    "",
    "## Equation",
    "",
    f"`{equation}`",
    "",
    "## Growth pathways",
    "",
]

lines.extend([f"- {item}" for item in growth_pathways])

lines.extend([
    "",
    "## Constraints to track",
    "",
])

lines.extend([f"- {item}" for item in constraints])

lines.extend([
    "",
    "## Notebook roadmap",
    "",
])

lines.extend([f"- **{nb}:** {purpose}" for nb, purpose in notebook_map])

lines.extend([
    "",
    "## Development questions",
    "",
])

lines.extend([f"{i}. {question}" for i, question in enumerate(questions, start=1)])

out = reports_dir / "context_summary.md"
out.write_text("\n".join(lines), encoding="utf-8")

print("Wrote:", out)
````
